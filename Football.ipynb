```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
football_df= pd.read_csv('/content/df_full_premierleague.csv')

# Manchester City stats home
manc_home= football_df[football_df['home_team']== 'Manchester City']
manc_home= manc_home.rename(columns={'goal_home_ft':'goals_scored'})
manc_home= manc_home.rename(columns={'goal_away_ft':'goals_conceded'})
manc_home= manc_home.rename(columns={'home_shots_on_target':'shots_on_target'})
manc_home= manc_home.rename(columns={'home_possession':'team_possession'})

# Manchester City stats away
manc_away= football_df[football_df['away_team']== 'Manchester City']
manc_away= manc_away.rename(columns={'goal_home_ft':'goals_conceded'})
manc_away= manc_away.rename(columns={'goal_away_ft':'goals_scored'})
manc_away= manc_away.rename(columns={'away_shots_on_target':'shots_on_target'})
manc_away= manc_away.rename(columns={'away_possession':'team_possession'})

# total stats
manc_all_games= pd.concat([manc_home,manc_away],axis=0)
manc_all_games['goal_difference']= manc_all_games['goals_scored']-manc_all_games['goals_conceded']
manc_all_games['Points']=0
manc_all_games.loc[manc_all_games['goal_difference']>0,'Points']=3
manc_all_games.loc[manc_all_games['goal_difference']==0,'Points']=1
manc_total= manc_all_games[['goals_scored','goals_conceded','goal_difference','shots_on_target','team_possession','Points']]
manc_all_games
```

```python
x_train,y_train= tot_total.iloc[:,:5], tot_total.iloc[:,-1]

#linear regression on selected parameters
from sklearn.linear_model import LinearRegression
sklearn_model= LinearRegression().fit(x_train,y_train)
sklearn_y_predictions = sklearn_model.predict(x_train)

#error calculation
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
mean_absolute_error(sklearn_y_predictions, y_train),mean_absolute_percentage_error(sklearn_y_predictions,y_train), sklearn_y_predictions
tot_total['sklearn_y_predictions'] = sklearn_y_predictions
tot_total
```
